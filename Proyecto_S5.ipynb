[Junta los datos agregados en un DataFrame para que haya un registro que represente lo que consumió un usuario único en un mes determinado.]

```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month
calls_duration_df = calls_duration.rename(columns={"call_date": "month"})
messages_amount_df = messages_amount.reset_index(name="total_messages").rename(columns={"message_date": "month"})
data_usage_df = data_usage.reset_index(name="total_mb").rename(columns={"session_date": "month"})

merged_data = (monthly_calls_df
    .merge(calls_duration_df, on=["user_id", "month"], how="outer")
    .merge(messages_amount_df, on=["user_id", "month"], how="outer")
    .merge(data_usage_df, on=["user_id", "month"], how="outer")
).fillna(0)

print(merged_data.head())
```

## Prueba las hipótesis estadísticas

[Prueba la hipótesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa.]

En este caso,nuestra hipótesis nula plantearía que el ingreso mensual promedio por usuario es el mismo en ambos planes; en otras palabras, el plan elegido no cambia el ingreso esperado. La hipótesis alternativa por otra parte establecería el hecho de que los ingresos promedio difieren entre planes. 

Para contrastarlas usaré una prueba t de dos muestras con corrección de Welch, adecuada cuando las varianzas no son iguales. En cuanto al nivel de significancia, pienso que regularmente se establece en .05; si el valor p resulta menor que 0.05, concluiré que existe evidencia de una diferencia en el ingreso promedio entre Surf y Ultimate, de lo contrario se mantendría la hipótesis nula y concluiríamos en que con estos datos no se puede rechazar la hipotesis nula. 

```python
# Prueba las hipótesis
rev_surf = final_data.loc[final_data["plan"] == "surf", "monthly_revenue"]
rev_ultimate = final_data.loc[final_data["plan"] == "ultimate", "monthly_revenue"]
alpha = 0.05
p = st.ttest_ind(rev_ultimate, rev_surf).pvalue  # prueba t de dos muestras, bilateral
print("p-value:", p)

if p < alpha:
    print("Conclusión: p < 0.05 → hay evidencia de diferencia en el ingreso promedio entre planes.")
else:
    print("Conclusión: p ≥ 0.05 → no se rechaza H0.")
```

[Prueba la hipótesis de que el ingreso promedio de los usuarios del área NY-NJ es diferente al de los usuarios de otras regiones.]

[Elabora las hipótesis nula y alternativa, escoge la prueba estadística, determina el valor alfa

La hipótesis nula establece que el ingreso mensual promedio por usuario en el área NY–NJ es el mismo que en el resto de las regiones, o sea que la región de residencia no modifica el ingreso esperado. La hipótesis alternativa sostiene que los ingresos promedio difieren entre NY–NJ y las demás regiones. Para contrastarlas aplicaré una prueba t de dos muestras con ttest_ind, en contraste bilateral y con un nivel de significancia de 0.05. Si el valor p es menor que 0.05, concluiré que existe evidencia de una diferencia en el ingreso promedio entre NY–NJ y otras regiones, pero si es mayor o igual que 0.05, no  se rechazará la hipótesis nula.

```python
# Prueba las hipótesis
nynj_ids = df_users.loc[df_users["city"] == "New York-Newark-Jersey City, NY-NJ-PA MSA","user_id"]

rev_nynj  = final_data.loc[ final_data["user_id"].isin(nynj_ids),  "monthly_revenue"]
rev_other = final_data.loc[~final_data["user_id"].isin(nynj_ids), "monthly_revenue"]

alpha = 0.05
p = st.ttest_ind(rev_nynj, rev_other).pvalue
print("p-value:", p)

if p < alpha:
    print("Conclusión: p < 0.05 → hay evidencia de diferencia en el ingreso promedio entre regiones.")
else:
    print("Conclusión: p ≥ 0.05 → no se rechaza H0.")
```

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es *estadísticamente diferente*. En futuros notebooks podrías utilizar este test para decidir sobre la igualdad de varianzas. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).
